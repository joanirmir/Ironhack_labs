### Cleaning and wrangling data

- Some car brand names refer to the same brand. Replace all brand names that contain the word "Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use `loc` with boolean indexing.

- Convert CO2 Emissions from Grams/Mile to Grams/Km

- Create a binary column that solely indicates if the transmission of a car is automatic or manual. Use `pandas.Series.str.startswith` and .

- convert MPG columns to km_per_liter

```python
#Replace all brand names that contain the word"Dutton" for simply "Dutton"
vehicles_df.loc[vehicles_df["Make"].str.contains('Dutton') == True, "Make"] = "Dutton"

#Show if it is replaced
vehicles_df["Make"][vehicles_df["Make"].str.contains('Dutton')]

#Clean the other names too
vehicles_df.loc[vehicles_df["Make"].str.contains('BMW') == True, "Make"] = "BMW"
vehicles_df.loc[vehicles_df["Make"].str.contains('Saleen') == True, "Make"] = "Saleen"
vehicles_df.loc[vehicles_df["Make"].str.contains('PAS') == True, "Make"] = "PAS"
vehicles_df.loc[vehicles_df["Make"].str.contains('Panos') == True, "Make"] = "Panoz Auto Development"

#Count the values of the column Make to see if it was replaced
print(vehicles_df["Make"].value_counts())
```

```python
#Convert CO2 Emissions from Grams/Mile to Grams/Km
print(vehicles_df.loc[:, "CO2 Emission Grams/Mile"])
vehicles_df.loc[:,"CO2 Emission Grams/Mile"] = vehicles_df.loc[:,"CO2 Emission Grams/Mile"] / 1.60934
```

```python
#Rename the column 'CO2 Emission Grams/Mile' to 'CO2 Emission Grams/Km'
vehicles_df.rename(columns = {"CO2 Emission Grams/Mile":"CO2 Emission Grams/Km"}, inplace = True)
print(vehicles_df.info)
```

```python
#The column was renamed
vehicles_df.info()
```

```python
#Create a binary column that solely indicates if the transmission of a car is automatic or manual. 
#Use pandas.Series.str.startswith
```

```python
#Converting MPG columns to km_per_liter
vehicles_df.loc[:,"City MPG"] = vehicles_df.loc[:,"City MPG"] * (1.60934/3.78541)

# Same with Higway MPG, Combined MPG

vehicles_df.loc[:,"Highway MPG"] = vehicles_df.loc[:,"Highway MPG"] * (1.60934/3.78541)

vehicles_df.loc[:,"Combined MPG"] = vehicles_df.loc[:,"Combined MPG"] * (1.60934/3.78541)

print(vehicles_df["City MPG"].value_counts())
```

Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Grams/Mile * Mile/Km -> Grams/Mile * 1 Mile/1.60934Km

$$ \frac{Grams}{Mile} * \frac{Mile}{Km} $$

$$ \frac{Grams}{Mile} * \frac{1 Mile}{1.60934Km}  $$

convert MPG columns to km_per_liter

MPG = Miles/Gallon -> Km/Liter

1 Mile = 1.60934 Km

1 Gallon = 3.78541 Liters

$$ \frac{Miles}{Gallon} -> \frac{Miles}{Gallon} * \frac{Km}{Miles} * \frac{Gallon}{Liters}$$

$$ \frac{Miles}{Gallon} -> \frac{Miles}{Gallon} * \frac{1.60934Km}{ 1Miles} * \frac{1 Gallon}{3.78541 Liters}$$

* ( 1.60934 / 3.78541 )
