#### Standarizing header names

Drop the index column first:

```python
def drop_columns(insurance_df):
    insurance_df.drop(columns=["Unnamed: 0"], inplace=True, axis=1)
    return insurance_df
```

```python
insurance_df 
```

```python
def lower_case_column_names(insurance_df):
    insurance_df.columns=[i. lower() for i in insurance_df.columns]
    return insurance_df
```

```python
insurance_df = lower_case_column_names(insurance_df)
```

```python
insurance_df
```

Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter, ie. January, February and March. Hint: If data from March does not exist, consider only January and February.

```python
def extract_months_and_create_a_new_column(insurance_df):
    insurance_df['month'] = pd.to_datetime(insurance_df['effective to date']).dt.month
    return insurance_df
```

```python
insurance_df = extract_months_and_create_a_new_column(insurance_df)
insurance_df
```

Convert integer to categorical data type:

```python
def convert_to_categorical_data(insurance_df):
    insurance_df['month'] = insurance_df['month'].astype("category")
    insurance_df['month'].cat.categories = ['january','feburary']
    return insurance_df
```

```python
insurance_df = convert_to_categorical_data(insurance_df)
```

Highly Correlated Columns

```python
def drop_high_correlated_columns(insurance_df, threshold):
    corr_matrix = insurance_df.corr().abs()
    upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    [column for column in upper_triangle.columns if any(upper_triangle[column]>threshold)]
```

Drop one of the two features that show a high correlation between them (greater than 0.9). If there is no pair of features that have a high correlation, then do not drop any features.

```python
drop_high_correlated_columns(insurance_df, 0.9)
```
