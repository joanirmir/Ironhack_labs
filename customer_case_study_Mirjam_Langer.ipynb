Import libraries and concatinate the csv_files first

```python
import numpy as np
import pandas as pd
```

#### Rename the columsn 'State' and 'Gender' in file3 and merge the files together

```python
file3_df = pd.read_csv('data/file3.csv')
```

```python
def rename_columns(file3_df):
    file3_df.rename(columns ={'State':'ST', 'Gender':'GENDER'}, inplace = True)
    return file3_df
```

```python
rename_columns(file3_df)
```

```python
def concatinate_data_files():
    file1 = pd.read_csv('data/file1.csv')
    file2 = pd.read_csv('data/file2.csv')
    file3 = file3_df
    return pd.concat([file1,file2,file3], axis=0)
```

```python
insurance_df = concatinate_data_files()
```

```python
insurance_df
```

#### Standarizing header names


First lower case column names:

```python
def lower_case_column_names(insurance_df):
    insurance_df.columns=[i. lower() for i in insurance_df.columns]
    return insurance_df
```

The columns 'customer lifetime value' and 'number of open complaints' have as data type object.
